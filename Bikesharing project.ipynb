```python
categorical_col = ['Hour','Temperature(C)','Humidity(%)','Wind speed (m/s)','Visibilty10m']
contineous_col = ['Dew point temperature(C)','Solar Radiation (MJ/m2)','Rainfall(mm)','Snowfall (cm)','Seasons','Holiday']
target = ['Rented Bike Count']
```

```python
for col in categorical_col:
    plt.figure(figsize=(4,3))
    sns.boxplot(bike[col],orient="h",width=0.8)
    plt.show()
```

```python
for col in contineous_col:
    sns.displot(bike[col],kde=True,edgecolor='white',height=4)
    plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.countplot(y="Hour",data=bike,palette="flare")
plt.ylabel('Hour')
plt.xlabel('Total')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.countplot(y="Seasons",data=bike,palette="rocket")
plt.ylabel('Seasons')
plt.xlabel('Total')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
sns.countplot(y="Holiday",data=bike,palette="rocket")
plt.ylabel('Holiday')
plt.xlabel('Total')
plt.show()
```
